## 3. main_df data exploration

```python
# Convert 'dteday' column to datetime format
day_df['dteday'] = pd.to_datetime(day_df['dteday'])
hour_df['dteday'] = pd.to_datetime(hour_df['dteday'])

# Merge day_df and hour_df
main_df = pd.merge(
    left=day_df,
    right=hour_df,
    how="left",
    left_on="dteday",
    right_on="dteday"
)

# Check the first 5 rows
main_df.head()
```

```python
# Table overview
main_df.info()
```

# Data Visualization
## Visualisasi tren penyewaan sepeda secara keseluruhan

```python
# Menggabungkan data berdasarkan bulan
monthly_df = main_df.resample('M', on='dteday').sum()

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(monthly_df.index, monthly_df['cnt_x'], color='#6499E9')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Penyewaan Sepeda')
plt.title('Penyewaan Sepeda Seiring Waktu (Digabung Berdasarkan Bulan)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Visualisasi penyewaan sepeda berdasarkan musim, bulan, dan hari
### Berdasarkan Musim

```python
# Menggabungkan data berdasarkan musim
season_agg = main_df.groupby("season_x").agg({
    "instant_x": "nunique",
    "cnt_x": ["max", "min"]
})

# Plot the data
plt.figure(figsize=(10, 6))
x = season_agg.index
y_max = season_agg[('cnt_x', 'max')]
y_min = season_agg[('cnt_x', 'min')]

plt.bar(x, y_max, label='Max Rentals', color='#6499E9')
plt.bar(x, y_min, label='Min Rentals', color='orange')

# Modify x-axis labels
season_labels = ['Musim ke-1', 'Musim ke-2', 'Musim ke-3', 'Musim ke-4']
plt.xticks(x, season_labels)

plt.xlabel('Musim')
plt.ylabel('Jumlah Penyewaan Sepeda')
plt.title('Penyewaan Sepeda Maks dan Min berdasarkan Musim')
plt.legend()

# Annotate the maximum and minimum values
for i, (max_val, min_val) in enumerate(zip(y_max, y_min)):
    plt.text(i+1, max_val, str(max_val), ha='center', va='bottom', fontweight='bold')
    plt.text(i+1, min_val, str(min_val), ha='center', va='bottom', fontweight='bold')
    
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Berdasarkan Bulan

```python
# Aggregate the data by month
monthly_agg = main_df.groupby("mnth_x").agg({
    "instant_x": "nunique",
    "cnt_x": ["max", "min"]
})

# Plot the data
plt.figure(figsize=(10, 6))
x = monthly_agg.index
y_max = monthly_agg[('cnt_x', 'max')]
y_min = monthly_agg[('cnt_x', 'min')]

plt.bar(x, y_max, label='Max Rentals', color='#6499E9')
plt.bar(x, y_min, label='Min Rentals', color='orange')

# Modify y-axis labels
month_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'Mei', 'Jun', 'Jul', 'Agu', 'Sep', 'Okt', 'Nov', 'Des']
plt.xticks(x,month_labels)

plt.xlabel('Bulan')
plt.ylabel('Jumlah Penyewaan Sepeda')
plt.title('Penyewaan Sepeda Maks dan Min per Bulan')
plt.legend()

# Annotate the maximum and minimum values
for i, (max_val, min_val) in enumerate(zip(y_max, y_min)):
    plt.text(i+1, max_val, str(max_val), ha='center', va='bottom', fontweight='bold')
    plt.text(i+1, min_val, str(min_val), ha='center', va='bottom', fontweight='bold')

plt.grid(True)
plt.tight_layout()
plt.show()
```

### Berdasarkan Hari

```python
# Aggregate the data by weekday
weekday_agg = main_df.groupby("weekday_x").agg({
    "instant_x": "nunique",
    "cnt_x": ["max", "min"]
})

# Plot the data
plt.figure(figsize=(10, 6))
x = weekday_agg.index
y_max = weekday_agg[('cnt_x', 'max')]
y_min = weekday_agg[('cnt_x', 'min')]

plt.bar(x, y_max, label='Max Rentals', color='#6499E9')
plt.bar(x, y_min, label='Min Rentals', color='orange')

plt.xlabel('Hari')
plt.ylabel('Jumlah Penyewaan Sepeda')
plt.title('Penyewaan Sepeda Maks dan Min pada Hari')
plt.xticks(rotation=0)
plt.legend()

# Annotate the maximum and minimum values
for i, (max_val, min_val) in enumerate(zip(y_max, y_min)):
    plt.text(i, max_val, str(max_val), ha='center', va='bottom', fontweight='bold')
    plt.text(i, min_val, str(min_val), ha='center', va='bottom', fontweight='bold')
    
# Adjust the position of the legend
plt.legend(loc='upper right', bbox_to_anchor=(1.15, 1))

plt.grid(True)
plt.tight_layout()
plt.show()
```

## Visualisasi jam penyewaan sepeda paling banyak dan paling sedikit

```python
# Aggregate the data by hour
hourly_agg = main_df.groupby("hr").agg({
    "instant_y": "nunique",
    "cnt_y": ["max", "min"]
})

# Plot the data
plt.figure(figsize=(10, 6))
x = hourly_agg.index
y_max = hourly_agg[('cnt_y', 'max')]
y_min = hourly_agg[('cnt_y', 'min')]

plt.bar(x, y_max, label='Max Rentals', color='#6499E9')
plt.bar(x, y_min, label='Min Rentals', color='orange')

plt.xlabel('Jam per Hari')
plt.ylabel('Jumlah Penyewaan Sepeda')
plt.title('Penyewaan Sepeda Maks dan Min per Jam')

# Modify x-axis labels
hour_labels = [str(i) for i in x]
plt.xticks(x, hour_labels)

plt.legend()

# Annotate the maximum and minimum values
for i, (max_val, min_val) in enumerate(zip(y_max, y_min)):
    plt.text(i, max_val, str(max_val), ha='center', va='bottom', fontweight='bold')
    plt.text(i, min_val, str(min_val), ha='center', va='bottom', fontweight='bold')

plt.grid(True)
plt.tight_layout()
plt.show()
```
